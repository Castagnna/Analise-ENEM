## importando dados de GeoLocalização

```python
df_gps = pd.read_csv(r'C:\PythonDin\analise-dados\DadosENEM\DADOS\municipiosGPS.csv', sep=',', encoding='iso-8859-1',\
                     usecols=[0,2,3])
```

```python
df_gps.head(3)
```

```python
df_gps.info()
```

### Agrupando DF

```python
df_merge = df_enem.merge(right=df_IDH, how='left', left_on='CO_MUNICIPIO_RESIDENCIA', right_on='Codmun7')
```

```python
df_merge.head(3)
```

```python
df_merge = df_merge.merge(right=df_gps, how='left', left_on='Codmun7', right_on='codigo_ibge')
```

```python
df_merge.head(3)
```

```python
df_merge.info()
```

```python
df_merge.drop('Codmun7', axis=1, inplace=True)
```

```python
# Convertendo os tipos de dados
df_merge.codigo_ibge = df_merge.codigo_ibge.astype('object')
```

```python
df_merge.info()
```

### Plotando

```python
# s= tamanho do pontp, c=referencia para escala de cores, cmap= padrãod e cores
df_merge.plot.scatter(y='latitude', x='longitude', figsize=(11,8), s=4,
                     c=df_merge.NU_NOTA_REDACAO,
                     cmap='RdYlGn',
                     title='Nota na Redação por municipio')
```

### Criando gráfico com subplots

```python
import matplotlib.pyplot as plt
```

```python
#Criando subplots
fig, ax = plt.subplots(2,2, figsize=(16,12))
s=3

# Preenchendo subplots
df_merge.plot.scatter(y='latitude', x='longitude', s=s,
                     c=df_merge.NU_NOTA_REDACAO,
                     cmap='RdYlGn',
                     title='Notas Redação por municipio',
                     ax=ax[0,0])

df_merge.plot.scatter(y='latitude', x='longitude', s=s,
                     c=df_merge.NU_NOTA_CN,
                     cmap='RdYlGn',
                     title='Notas Ciências por municipio',
                     ax=ax[0,1])

df_merge.plot.scatter(y='latitude', x='longitude', s=s,
                     c=df_merge.NU_NOTA_CH,
                     cmap='RdYlGn',
                     title='Notas Humanas por municipio',
                     ax=ax[1,0])

df_merge.plot.scatter(y='latitude', x='longitude', s=s,
                     c=df_merge.NU_NOTA_MT,
                     cmap='RdYlGn',
                     title='Nota Matemática por municipio',
                     ax=ax[1,1])
```
